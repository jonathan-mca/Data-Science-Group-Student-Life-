```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio
pio.renderers.defaule = 'colab'
```

```python
url = 'https://raw.githubusercontent.com/lkp04/Data-File/refs/heads/main/ResearchInformation3.csv'
df = pd.read_csv(url)
df
```

```python
df_copy1 = df.drop(['SSC'], axis=1)
df_copy1
df_copy2 = df.drop(['SSC','Semester','Computer','Job','Department'], axis=1)
df_copy2
```

```python
#df_copy1.describe()
df_copy1.dtypes
#df_copy1.describe(include='all')

#df_copy1['Overall'] = df_copy1['Overall'].astype('float64')
df_copy2['English'] = df_copy2['English'].astype('int64')


df_copy2.dtypes
#df_copy2
```

```python
df_copy2.describe()
```

```python
df_copy2.value_counts()
```

```python
df2 = df_copy2.groupby('English')['Overall'].mean().reset_index()
df2
fig = px.bar(df_copy2,
                y='English',
                x='Overall',
                title='grade vs. English Compreshension',
              color = 'Gender')
fig.show()
```

```python
fig = px.box(df_copy2,x='Overall', y='Gaming', color ="Gender")

fig.update_layout(
                  xaxis_title="Overall Grade",
                  yaxis_title="Hours spent Gaming"
                  )


fig.show()
```

```python

fig = px.scatter(df_copy2,
                y='HSC',
                x='Overall',
                title='High school Grade vs. Overall Grade in College',
                color = 'Gender'
                )

fig.show()
```
