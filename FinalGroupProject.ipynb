#Proposed Questions
 In this paper, we will be looking into these four Data sets and exploring two central Questions: 1. Are there any identifiable correlations between students habits and their Overall GPA? and 2. Do these relationships differ regionally or are patterns generally the same?

 To answer these questions, First we want to find a couple variables that all four data sets have in common to create a meaningful comparsison. After establishing these variables we will be creating our plots or graphs to help examine trends within each data set.
```python
fig = px.box(df_copy2,x='Overall', y='Gaming', color ="Gender", title='Hours spent gaming vs. Overall GPA')

fig.update_layout(
                  xaxis_title="Overall Grade",
                  yaxis_title="Hours spent Gaming"
                  )


fig.show()
```
